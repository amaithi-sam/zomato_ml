### Feature: location

https://www.analyticsvidhya.com/blog/2015/11/easy-methods-deal-categorical-variables-predictive-modeling/?utm_source=blog&utm_medium=Categorical_data_encoding

Combine levels: To avoid redundant levels in a categorical variable and to deal with rare levels, we can simply combine the different levels. There are various methods of combining levels. Here are commonly used ones:

    Using Business Logic: It is one of the most effective method of combining levels. It makes sense also to combine similar levels into similar groups based on domain or business experience. For example, we can combine levels of a variable “zip code” at state or district level. This will reduce the number of levels and improve the model performance also.

```python
from geopy.geocoders import Nominatim
from pprint import pprint
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

For a structured query, provide a dictionary whose keys
    are one of: `street`, `city`, `county`, `state`, `country`, or
    `postalcode`

```python
locat = {'Ejipura, bangalore, india'}

location = geolocator.geocode(locat)
pprint(location.raw)
```

```python
d = location.raw
ld = d['display_name'].split(',')
print(int(ld[-2]))
```

```python
def pincode_finder(location_in):

    locat = f'{location_in}, bangalore, india'
    location = geolocator.geocode(locat)
    d = location.raw
    ld = d['display_name'].split(', ')
    if ld[-2] == 'Karnataka':
        ld[-2] = np.nan
    return location_in, ld[-2]
```

```python
loc = {}
for locat in df['location'].unique().tolist():
    try:
        lc, code = pincode_finder(locat)
    except:
        lc = locat 
        code = np.nan 
    
    loc[lc] = code
    
pprint(loc)
```
