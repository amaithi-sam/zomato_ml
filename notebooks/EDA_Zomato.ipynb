#### Handle the cost2people variable

```python
def handle_cost2people(val):
    val = str(val)
    if ',' in val:
        val = val.replace(',','')
        return float(val)
    else:
        return float(val)
    
df['cost2people'] = df['cost2people'].apply(handle_cost2people)
df.head(3)
```

### creating new feature 'cost_per_person'

```python
df['cost_per_person'] = df['cost2people']/2
```

### Handling location Column

```python
locat = df['location'].value_counts(ascending=False)
locat_less_50 = locat[locat<50]

def correct_locat(value):
    if (value in locat_less_50):
        return 'others'
    else:
        return value 

df['location'] = df['location'].apply(correct_locat)
df['location'].value_counts()
```

### Cleaning the rest type column

```python
rest_type = df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type = df['rest_type'].unique().tolist()
```

```python
rest_type
```

```python
# Remove Duplicates in the rest_type list

new_type = []

for type in rest_type:
    if ',' in type:
        new_type.append(type.split(',')[0])
        new_type.append(type.split(', ')[1])
    else:
        new_type.append(type)
        
new_type
        
        
```

```python
len(new_type)
```

```python
# Remove Duplicates in the new_type list
rest_type_ = [*set(new_type)]
rest_type_, len(rest_type_)
```

```python
for val in new_type:
    if val in rest_type_:
        print('True')
    else:
        print('False')
```

##### Saving the unique Rest_types in csv

```python
df_type = pd.DataFrame(rest_type_, columns=['rest_type'])
```

```python
df_type.to_csv('rest_type_unique.csv')
```

### Handling cuisines Column

```python
df['cuisines'].value_counts()
```

```python
cui_list = df['cuisines'].unique().tolist()
```

```python
cui_list
```

```python
# Splitting Cuisine 

cui_ls = []

for cui in cui_list:
    if ',' in cui:
        for one_cui in cui.split(','):
            try:
                cui_ls.append(one_cui.replace(' ', '', 1))
            except:
                cui_ls.append(one_cui)
        
    else:
        cui_ls.append(cui)
        
cui_ls
        
        
```

```python
# REmoving the Duplicates in the Cuisines
final_cuisine = [*set(cui_ls)]
final_cuisine, len(final_cuisine)
```

```python
df_cui = pd.DataFrame(final_cuisine, columns=['cuisines'])
df_cui.to_csv('rest_cuisines_unique.csv')
```
